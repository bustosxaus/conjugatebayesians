```python
purchaset1 = purchaset
```

```python
purchaset1.dtypes
```

```python
obj_cols = purchaset1.select_dtypes(['object']).columns
```

```python
purchaset1[obj_cols] = purchaset1[obj_cols].apply(lambda x: x.astype('category'))
```

```python
purchaset1.dtypes
```

```python
cat_cols = purchaset1.select_dtypes(['category']).columns
```

```python
purchaset1[cat_cols] = purchaset1[cat_cols].apply(lambda x: x.cat.codes)
```

```python
purchaset1.dtypes
```

```python
heatmap = plt.pcolor(purchaset1.corr(), cmap='jet')
plt.colorbar(heatmap)
plt.gca().set_aspect('equal')
```

```python
kmeans.fit(purchaset1)
```
